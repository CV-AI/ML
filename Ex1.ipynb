```python
# import sth
import numpy as np
import pandas as pd
import pickle
import sys
input_file = "Lab01-Data/lab1_Ex1_data.csv"
data_frame = pd.read_csv(input_file)
print(data_frame.describe())
output_dict = {}
```

```python
output_dict['Q1'] = dict(data_frame.max(numeric_only=True))
print(output_dict['Q1'])
```

```python
output_dict['Q2'] = dict(data_frame.mean(numeric_only=True))
print(output_dict['Q2'])
```

```python
print(data_frame['Sex'])
sex_type = []
for row in data_frame.itertuples():
    if getattr(row, 'Sex') not in sex_type:
        sex_type.append(getattr(row, 'Sex'))
print(sex_type)
output_dict['Q3'] = len(sex_type)
```

```python
print(output_dict)
```

```python
output_dict['Q4'] = data_frame.isna().sum().sum() # sum twice to add different column together
print(output_dict['Q4'])
```

```python
data_frame_copy = data_frame.copy()
print(data_frame.median(numeric_only=True)['Age'])
data_frame_copy.at[:, 'Age'] = data_frame.median(numeric_only=True)['Age']
output_dict['Q5'] = data_frame_copy
print(data_frame_copy)
```

```python
df_copy_Q6 = data_frame.copy()
df_copy_Q6 = df_copy_Q6.where(df_copy_Q6['Sex']=='male').dropna(how='all') # get all male data, how='all' means only drop when all data is NaN
df_copy_Q6 = df_copy_Q6.sort_values(by='PassengerId', ascending=True)
print(df_copy_Q6)
output_dict['Q6'] = df_copy_Q6
```

```python
df_copy_Q7 = data_frame.copy()
df_copy_Q7 = df_copy_Q7.where(df_copy_Q7['Sex']=='male').dropna(how='all') # get all male data
df_copy_Q7 = df_copy_Q7.where(df_copy_Q7['Age']>20).dropna(how='all') # get people whose age is larger that 20
df_copy_Q7 = df_copy_Q7.sort_values(by='PassengerId', ascending=True)
print(df_copy_Q7)
output_dict['Q7'] = df_copy_Q7
```

```python
df_copy_Q8 = data_frame.copy()
df_copy_Q8['NewSex'] = df_copy_Q8['Sex']
df_copy_Q8['NewSex'] = df_copy_Q8['NewSex'].map({'female':2, 'male':1})
df_copy_Q8 = df_copy_Q8.sort_values(by='PassengerId', ascending=True)
print(df_copy_Q8)
```

```python
df_copy_Q9 = data_frame.copy()
df_copy_Q9['nE1'] = df_copy_Q9['Embarked']
df_copy_Q9['nE2'] = df_copy_Q9['Embarked']
df_copy_Q9['nE3'] = df_copy_Q9['Embarked']
df_copy_Q9['nE1'] = df_copy_Q9['nE1'].map({'Q': 1, 'S':0, 'C':0})
df_copy_Q9['nE2'] = df_copy_Q9['nE2'].map({'Q': 0, 'S':1, 'C':0})
df_copy_Q9['nE3'] = df_copy_Q9['nE3'].map({'Q': 0, 'S':0, 'C':1})
df_copy_Q8 = df_copy_Q8.sort_values(by='PassengerId', ascending=True)
print(df_copy_Q9)
output_dict['Q9'] = df_copy_Q9
```

```python
with open("Lab1_Ex1.pkl", 'wb') as myfile:
    pickle.dump(output_dict, myfile)
```
